```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import pandas as pd
```

```python
# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')
```

```python
driver_path = r"C:\Users\ANZONY\Downloads\chromedriver.exe"


driver = webdriver.Chrome(driver_path, chrome_options=options)

# Inicializamos el navegador
driver.get('https://www.convocatoriascas.com')
sleep(1)

convocatorias = driver.find_elements_by_xpath('//article[@class="post"]')
titulos_total = []
detalles_total = []

for convocatoria in convocatorias:
    # Aca appendeo los titulos
    titulo = convocatoria.find_element_by_xpath('.//h2[@class="titulo"]')
    titulo_1 = titulo.text
    titulos_total.append(titulo_1)
    print(titulo_1)
    
    # Aca apendeo los detalles
    detalles = convocatoria.find_element_by_xpath('.//div[@class="detalles"]')
    detalles_1 = detalles.text
    detalles_total.append(detalles_1)
    print(detalles_1)
    
```

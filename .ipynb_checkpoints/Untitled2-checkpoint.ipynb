```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")
```

```python
driver_path = r"C:\Users\ANZONY\Downloads\chromedriver.exe"
driver = webdriver.Chrome(driver_path , options =options)
driver.get('https://www.convocatoriasdetrabajo.com/convocatorias-cas-vigentes.php')
time.sleep(1)
```

```python
i = 1
caja ="vacia"
convocatorias = []
fechas = []
lugares = []
sueldos = []
```

```python
while caja !=0 :
    
            datas = driver.find_elements(By.XPATH,f'/html/body/section/section/article[{i}]')
    
            # Convocatoria
            convocatoria = datas[0].find_element(By.XPATH,f'/html/body/section/section/article[{i}]/h4/a').text
            match1 = re.search(r'.*:\s(.*)',convocatoria)
            match2 = re.search(r'.*:*^([0-9]+.*)',convocatoria)
            match3 = re.search(r'[0-9]+.*',convocatoria)
            
            if match1 :
                conv = match1.group(1)
            if match2 :
                conv = match2.group(1)
            if match3 :   
                conv = match3.group(0)
            else:
                conv = ""
        
            convocatorias.append(conv)
            print(conv)
            
            # fechas
            fecha = datas[0].find_element(By.XPATH,f'/html/body/section/section/article[{i}]/div[2]/div[1]').text
            fechas.append(fecha)
            print(fecha)
            
            #lugares
            lugar = datas[0].find_element(By.XPATH,f'/html/body/section/section/article[{i}]/div[2]/div[2]').text
            lugares.append(lugar)
            print(lugar)
    
           
            #sueldo
            sueldo = datas[0].find_element(By.XPATH,f'/html/body/section/section/article[{i}]/div[2]/div[3]').text
            sueldos.append(sueldo)
            print(sueldo)
        
        
            i += 1
        
             # Para que rompa cadena
            datas =  driver.find_elements(By.XPATH,f'/html/body/section/section/article[{i}]')
            caja = len(datas)                             
                                              
   
    
   
    
                                          
    
                            
    
            
    
    
    
```

```python
data_salario = {"Convocatoria" : convocatorias, "Fecha" : fechas , "Lugar" : lugares , "Sueldo": sueldos}
```

```python
data_pre_profesional = pd.DataFrame( data_salario )
```

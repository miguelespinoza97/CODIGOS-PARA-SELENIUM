```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")
```

```python
driver_path = r"C:\Users\User\Desktop\PRACTICAS MIGUEL\pyton\chromedriver.exe"
driver = webdriver.Chrome(driver_path , options =options)
driver.get('https://www.convocatoriascas.com/')
time.sleep(1)
```

```python
driver.find_element(By.XPATH, f'./html/body/section[1]/section/article[300]')
```

```python
i = 1
convocatorias = []
entidades = []
departamentos = []
remuneraciones = []

while  i > 0:
    
    datas=driver.find_elements(By.XPATH, f'/html/body/section[1]/section/article[{i}]')
    for data in datas:
        
        convocatoria=data.find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/h2/a').text
        match = re.search(r".*:\s(.*)", convocatoria)
        if match:
            conv = match.group(1)
        else:
            conv = ""
        print(conv)
        convocatorias.append( conv )
        
        entidad=data.find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[1]').text
        match = re.search(r".*:\s(.*)", entidad)
        if match:
            ent = match.group(1)
        else:
            ent = ""
        print(ent)
        
        
        entidades.append( ent )
        
        departamento =data.find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[2]').text
        match = re.search(r".*:\s(.*)", departamento)
        if match:
            dpt = match.group(1)
        else:
            dpt = ""
        
        print( departamento )
        departamentos.append( dpt )        
        
        remuneracion =data.find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[3]').text
        match = re.search(r".*:\s(.*)", remuneracion)
        if match:
            rem = match.group(1)
        else:
            rem = ""
        
        print( remuneracion )
        remuneraciones.append( rem )
    
    i += 1
    
    try:
        driver.find_element(By.XPATH, f'./html/body/section[1]/section/article[300]')
        print("hola")
    except:
        i == 0
        print("hola2")
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")
```

```python
driver_path = r"C:\Users\ANZONY\Downloads\chromedriver.exe"
driver = webdriver.Chrome(driver_path , options =options)
driver.get('https://www.convocatoriascas.com/')
time.sleep(1)
```

```python
i = 1
size_datas = 'no es vacio'
convocatorias = []
entidades = []
departamentos = []
remuneraciones = []

while  size_datas != 0 :
    
    datas = driver.find_elements(By.XPATH, f'/html/body/section[1]/section/article[{i}]')
     
    # Convocatoria
    convocatoria=datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/h2/a').text
    match = re.search(r".*:*\s^([0-9]+.*)", convocatoria)
    match2 = re.search(r".*:\s(.*)", convocatoria)
    match3 = re.search(r"[0-9]+.*", convocatoria)
    if match:
        conv = match.group(1)
    elif match2:
        conv = match2.group(1)
    elif match3:
        conv = match3.group(0)
    else:
        conv = ""
    print(conv)
  
    convocatorias.append( conv )

    # Entidades
    entidad=datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[1]').text
    match = re.search(r".*:\s(.*)", entidad)
    if match:
        ent = match.group(1)
    else:
        ent = ""
    print(ent)
    entidades.append( ent )
    
    # Departamentos
    departamento =datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[2]').text
    match = re.search(r".*:\s(.*)", departamento)
    if match:
        dpt = match.group(1)
    else:
        dpt = ""

    print( dpt )
  
    departamentos.append( dpt )        

    # Remuneraciones
    remuneracion =datas[0].find_element(By.XPATH, f'/html/body/section[1]/section/article[{i}]/div[2]/div/p[3]').text
    match = re.search(r".*:\s(.*)", remuneracion)
    if match:
        rem = match.group(1)
    else:
        rem = ""

    print( rem )
    remuneraciones.append( rem )

    i + = 1
   
    # Checking
    datas = driver.find_elements(By.XPATH, f'/html/body/section[1]/section/article[{i}]')
    size_datas = len( datas )
```

```python
data_salario = {"Convocatoria" : convocatorias, "Entidad" : entidades , "Departamentos" : departamentos , "Remuneracion": remuneraciones}
```

```python
data_pre_profesional = pd.DataFrame( data_salario )
```
